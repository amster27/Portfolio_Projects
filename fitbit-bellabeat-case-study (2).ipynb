### Sleep exploration

```r
#check sleep numbers
ggplot(sleep_day, aes(x=total_minutes_asleep, y=total_time_in_bed)) + geom_line()
head(sleep_day)

#check the average amount of minutes slept
sleep_day %>%
summarize(mean(total_minutes_asleep))
#average amount of sleep among users is 7 hours of sleep per night
sleep_day %>%
summarize(median(total_minutes_asleep))
#median amount of sleep among users is 7.22 hours of sleep per night
ggplot(sleep_day, aes(y=total_minutes_asleep)) + geom_boxplot()

#check the average amount of time users spend in bed, but not asleep
mean(sleep_day$total_time_in_bed) - mean(sleep_day$total_minutes_asleep)
#users spend an average of ~39 minutes in bed, but not asleep
#it might be good to track more information about when the lack of sleep is happening - i.e. before sleeping at night, after waking up in the morning, in the middle of the night
#additionally, it would be good to have records that show the exact time when users fall asleep

#show each user sleep patterns on different grids
ggplot(sleep_day, aes(x=activity_datetime, y=total_time_in_bed, group="id")) + geom_point() + facet_wrap("id")
ggplot(sleep_day, aes(x=activity_datetime, y=total_minutes_asleep, group="id")) + geom_point() + facet_wrap("id")
#8 users have 5 or less tracked sleep records, and 1 additional user only has 9 tracked sleep records
#With the versatility of Bellabeat, as opposed to Fitbit, it might be good to advertise that the device can be worn in different ways, and since it's a battery, you can easily track sleep records at night, because no charging necessary
```
